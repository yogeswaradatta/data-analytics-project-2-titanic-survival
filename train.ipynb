```python
import pandas as pd
df = pd.read_csv(r"D:\PYTHON_FILES\train\train.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(
    df.groupby(['Sex', 'Pclass'])['Age'].transform('median')
)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df['Cabin'] = df['Cabin'].fillna('No Data')
```

```python
df.isnull().sum()
```

```python
df['Embarked'].unique()
```

```python
df['Pclass'].unique()
```

```python
df['Embarked']= df['Embarked'].replace({'C':'Cherbourg',
                                        'Q': 'Queenstown',
                                         'S': 'Southampton'})
df['Pclass']= df['Pclass'].replace({1:'First Class',
                                        2: 'Second Class',
                                         3: 'Third Class'})
```

```python
df.head()
```

```python
df.drop(columns='PassengerId',inplace=True)
```

```python
df.head()
```

```python
df.insert(
    loc=df.columns.get_loc('Parch') + 1,  # position right after Parch
    column='FamilyType',
    value=(df['SibSp'] + df['Parch'] + 1).apply(
        lambda x: 'Alone' if x == 1 else ('Small Family' if x <= 4 else 'Large Family')
    )
)
```
